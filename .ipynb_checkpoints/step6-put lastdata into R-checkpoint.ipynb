```python
import pandas as pd
```

```python
Data = pd.read_csv("../data/last_data_svm_rfe.csv", sep=',',encoding = 'utf-8')
Data = Data.rename(columns={"Unnamed: 0":"Sample"})
Data.set_index(["Sample"], inplace=True)
Data.sort_values(by="Group" , inplace=True, ascending=True) 
Data
```

```python
msi = Data.iloc[:,0]
msi = pd.DataFrame(msi)
msi
```

```python
# msi.to_csv("./COAD/final_msi.txt", sep='\t',encoding = 'utf-8')
msi.to_csv("./R_code/data/final_msi.csv", sep=',',encoding = 'utf-8')
```
