```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
import re
from sklearn.feature_selection import SequentialFeatureSelector
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from scipy import stats
from statsmodels.stats.multitest import fdrcorrection
from boruta import BorutaPy
from sklearn.ensemble import RandomForestClassifier as RFC
from time import time
import datetime
from sklearn.metrics import roc_auc_score, recall_score,roc_curve
from sklearn.metrics import confusion_matrix as CM, recall_score as R
```

```python
new_data = pd.read_csv("../data/All_FilterData.csv", sep=',',index_col=0,encoding = 'utf-8')
new_data
```

```python
# Transpose new_data
new_data = pd.DataFrame(new_data.values.T, index=new_data.columns, columns=new_data.index)
new_data
```

```python
l = new_data.shape[1]
dic = {}  # Stores the mapping between variance and index
for i in range(l):
    dic[i] = new_data.iloc[:,i].var()  # Traverse to calculate the variance of each column, skipna defaults to true.
dic
```

```python
x = [i[0] for i in sorted(dic.items(), reverse=True, key=lambda item: item[1])]  # Sort dictionary value, return a list of tuples
x
```

```python
# Pick about one percent of the features
new_data1 = new_data.iloc[:,x[0:int(0.01 * l)]]
new_data1
```

```python
# In addition to obtaining sample data, the corresponding msi label is also required, 
# and the sample data needs to be matched with the corresponding label
```

```python
# find samples with msi
data_MSI = pd.read_csv("../data/COAD_data_MSI.csv",sep = ',',encoding = 'utf-8',usecols=['simple_tumor4','Predicted_MSI_Status'])
data_MSI
```

```python
# Convert the above dataframe into a list format to use regular expressions to take out the sample names and
data_MSI_list = data_MSI.values.tolist()
len(data_MSI_list)
data_MSI_list
```

```python
# First obtain the tuple corresponding to the sample label and msi, and then sort according to the current label data
dif_MSI_tuple = []
for i in new_data.index:
    for j in data_MSI_list:
        if i == j[1]:
            dif_MSI_tuple.append(j)
            break
```

```python
len(dif_MSI_tuple)
dif_MSI_tuple
```

```python
sample_rowname = new_data1._stat_axis.values.tolist() # row name
sample_listname = new_data1.columns.values.tolist()    # column name
```

```python
# put in order
sample_msi_char = []
for i in sample_rowname:
    for j in dif_MSI_tuple:
        if i == j[1]:
            sample_msi_char.append(j[0])
            break
```

```python
len(sample_msi_char)
sample_msi_char
```

```python
# Change msi to 1, mss to 0
sample_msi_num = []
for i in sample_msi_char:
    if i == "MSS":
        sample_msi_num.append(0)
    else:
        sample_msi_num.append(1)
```

```python
len(sample_msi_char)
# sample_msi_num
```

```python
# Insert the MSI of each sample into it
new_data1.insert(0,"Group",sample_msi_num)
new_data1
```

```python
column_uniques = new_data1['Group'].unique()
column_uniques
temp_data1 = new_data1[new_data1['Group'].isin([np.int64(0)])]
temp_data2 = new_data1[new_data1['Group'].isin([np.int64(1)])]
temp_data1
temp_data2
```

```python
temp_data1.drop(columns = ['Group'],inplace = True)
temp_data2.drop(columns = ['Group'],inplace = True)
temp_data1
temp_data2
```

```python
# Then you can get the previous data, or you can get the data labels and data directly from here
# get msi and data separately
msi = new_data1.iloc[:,0]
final_data = new_data1.iloc[:,1:]
```

```python
final_data
```

```python
msi
```

```python
# Check the 0:mss, 1:msi ratio in msi
msi.value_counts() # 241:55
```

```python
statistic,p_value = stats.ttest_ind(temp_data1,temp_data2,equal_var = False)
len(p_value)
```

```python
(p_value < 0.05).sum()
```

```python
boolean,fdr = fdrcorrection(p_value,alpha = 0.05,method = 'indep',is_sorted=False)
len(fdr)
```

```python
(fdr < 0.05).sum()
```

```python
# Find the features in final_data with p-value less than 0.05
i = 0
index = []
for p in fdr:
    if p < 0.05:
        index.append(i)
    i = i + 1
len(index)
```

```python
new_data1 = final_data.iloc[:,index]
new_data1
```

```python
# Ok, now you get the data you want, and then process the data
# The data has been processed, and then dimensionality reduction and feature selection are performed.
# First use the packing method to sort the features
```

```python
tree_range = range(0,100,1)
accuracyall = []
recallall = []
aucall = []
scoreall = []
for i in tree_range:
    times = time()
    rfc = RFC(n_estimators=i+1,n_jobs=-1, class_weight='balanced',random_state=420)
    accuracy = cross_val_score(rfc,Xtrain2,Ytrain,scoring="balanced_accuracy",cv=StratifiedKFold(10)).mean()
    accuracyall.append(accuracy)
    score = cross_val_score(rfc,Xtrain2,Ytrain,scoring="f1",cv=StratifiedKFold(10)).mean()
    scoreall.append(score)
    recall = cross_val_score(rfc,Xtrain2,Ytrain,scoring="recall",cv=StratifiedKFold(10)).mean()
    recallall.append(recall)
    auc = cross_val_score(rfc,Xtrain2,Ytrain,scoring="roc_auc",cv=StratifiedKFold(10)).mean()
    aucall.append(auc)
    print("i= %f, testing accuracy is %f,recall is %f', auc is %f" %(i+1,accuracy,recall,auc))
    print(datetime.datetime.fromtimestamp(time()-times).strftime("%M:%S:%f"))
```

```python
print(max(scoreall),(scoreall.index(max(scoreall))*10)+1)
plt.figure()
plt.xlabel("n_estimators",fontsize=14)
plt.ylabel("score",fontsize=14)
plt.plot(tree_range,accuracyall,c="blue",label="accuracy")
plt.plot(tree_range,recallall,c="red",label="recall")
plt.plot(tree_range,aucall,c="black",label="auc")
# plt.plot(tree_range,scoreall,c="orange",label="score")
plt.legend(loc='lower right',fontsize=15)
plt.title("learning curve of RFC model",fontsize=19,loc='center',color='black',fontweight='medium')
plt.savefig("./pic/Learning_curve(n_estimators).jpg",dpi=300)
plt.show()
```

```python
times = time()
clf = RFC(n_estimators=29,n_jobs=-1, class_weight='balanced',random_state=420).fit(Xtrain2, Ytrain)
result = clf.predict(Xtest1)
print(result)
score = clf.score(Xtest1,Ytest)
recall = recall_score(Ytest, result)
auc = roc_auc_score(Ytest,result)

# print("testing accuracy is %f,recall is %f'" %(score,recall))
print("testing accuracy is %f,recall is %f', auc is %f" %(score,recall,auc))
print(datetime.datetime.fromtimestamp(time()-times).strftime("%M:%S:%f"))
```

```python
#encode the label
from sklearn.preprocessing import LabelEncoder # Label only
encorder = LabelEncoder().fit(Ytest)

Ytest1 = pd.DataFrame(encorder.transform(Ytest))
Ytest1 = Ytest1.iloc[:,0].ravel()
Ytest1 
```

```python
clf_proba = RFC(n_estimators=29,n_jobs=-1, class_weight='balanced',random_state=420).fit(Xtrain2, Ytrain)
probrange = np.linspace(clf_proba.predict_proba(Xtrain2)[:,1].min(),clf_proba.predict_proba(Xtrain2)[:,1].max(),num=50,endpoint=False)
```

```python
fpr1, tpr1, thersholds1 = roc_curve(Ytrain, clf_proba.predict(Xtrain2), pos_label=1)
fpr2, tpr2, thersholds2 = roc_curve(Ytest, clf_proba.predict(Xtest1), pos_label=1)
AUC1 = roc_auc_score(Ytrain,clf_proba.predict(Xtrain2))
AUC2 = roc_auc_score(Ytest,clf_proba.predict(Xtest1))
```

```python
plt.xlabel("Specificity",fontsize=14)
plt.ylabel("Sensitivity",fontsize=14)
plt.plot(fpr1,tpr1,color="red",lw=1,label='AUC in training cohort(AUC=%0.4f)'%AUC1)
plt.plot(fpr2,tpr2,color="green",lw=1,label='AUC in validation cohort(AUC=%0.4f)'%AUC2)
plt.plot(probrange+0.05,probrange+0.05,c="black",lw=1,linestyle="--")
plt.legend(loc='best',fontsize=11)
plt.title("The ROC of RFC model",fontsize=19,loc='center',color='black')
plt.savefig("./pic/ROC(RFC).jpg",dpi=300)
plt.show()
```
