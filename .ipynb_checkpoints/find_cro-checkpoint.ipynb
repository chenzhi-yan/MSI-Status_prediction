```python
# Change msi to 1, mss to 0
sample_msi_num = []
for i in Group:
    if i == "MSS":
        sample_msi_num.append(0)
    else:
        sample_msi_num.append(1)
len(sample_msi_num)
sample_msi_num

# Insert the MSI of each sample into it
cg_gene = pd.DataFrame(cg_gene.values.T, index=cg_gene.columns, columns=cg_gene.index)
cg_gene.insert(0,"Group",sample_msi_num)
cg_gene

column_uniques = cg_gene['Group'].unique()
column_uniques
temp_data1 = cg_gene[cg_gene['Group'].isin([np.int64(0)])]
temp_data2 = cg_gene[cg_gene['Group'].isin([np.int64(1)])]
temp_data1
temp_data2

temp_data1.drop(columns = ['Group'],inplace = True)
temp_data2.drop(columns = ['Group'],inplace = True)
temp_data1
temp_data2

from scipy import stats
from statsmodels.stats.multitest import fdrcorrection
statistic,p_value = stats.ttest_ind(temp_data1,temp_data2,equal_var = False)
len(p_value)

(p_value < 0.05).sum()

boolean,fdr = fdrcorrection(p_value,alpha = 0.05,method = 'indep',is_sorted=False)
len(fdr)
fdr

(fdr < 0.05).sum()
```
